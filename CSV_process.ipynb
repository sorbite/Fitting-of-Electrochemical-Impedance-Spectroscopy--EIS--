```python
import numpy as np
data = np.genfromtxt('./EIS_data.csv', delimiter=',')
for i in range(0,12):
    csv = data[:,[3*i,3*i+1,3*i+2]]
    fname = 'csv' + str(i) + '.txt'
    print(csv)
    np.savetxt(fname, csv, fmt='%.18e', delimiter=' ', newline='\n', header='', footer='', comments='# ', encoding=None)
```
